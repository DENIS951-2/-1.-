# 13. У какого режиссера самые большие суммарные кассовые сбооры?

```python
answers['13'] = 'Peter Jackson'
```

```python
data = pd.read_csv('movie_bd_v5.csv')
def director_revenue(df):
    summ = Counter()
    for i in range(len(df)):
        for j in df.iloc[i].director.split('|'):
            summ[j] += df.iloc[i].revenue
    return pd.DataFrame.from_dict(summ, orient='index', columns = ['Summ'])
display(director_revenue(data[data.revenue > 0]).sort_values(by=['Summ'], ascending = False))
```

# 23. Для какого режиссера зима – самое продуктивное время года? 

```python
answers['23'] = 'Peter Jackson'
```

```python
df = data.copy()
def month_to_season(month):
    seasons = {
        12: 'winter',
        1: 'winter',
        2: 'winter',
        3: 'spring',
        4: 'spring',
        5: 'spring',
        6: 'summer',
        7: 'summer',
        8: 'summer',
        9: 'autumn',
        10: 'autumn',
        11: 'autumn',
    }
    return seasons[month]
df['release_month'] = pd.to_datetime(df.release_date).apply(lambda x: x.month)
df['season'] = df.release_month.apply(month_to_season)
def directors(row):
    for director in row.director.split('|'):
        return director, row.season
df = pd.DataFrame(df.apply(directors, axis=1).tolist(), columns=['director', 'season'])
df['count'] = 1
df = pd.pivot_table(df, index='director', columns='season', values='count', aggfunc='sum')
df.sort_values('winter', ascending=False).head(1)
```

# 24. Какая студия дает самые длинные названия своим фильмам по количеству символов?

```python
answers['24'] = 'Four By Two Productions'
```

```python
s_companies = data['production_companies'].str.split('|',expand=True).stack().unique()
df_result = pd.DataFrame(columns = ['production_companies', 'length'])
df_result['length'] = df_result['length'].astype('int')
for company in s_companies:
    df_company = data[data['production_companies'].str.contains(company,False)]['original_title']
    for tagline in df_company:
        index = len(df_result)
        df_result.loc[index,'production_companies'] = company
        df_result.loc[index,'length']  = len(tagline)
df_result.groupby('production_companies')['length'].mean().sort_values(ascending=False).head(1)
```

# 25. Описание фильмов какой студии в среднем самые длинные по количеству слов?

```python
answers['25'] = 'Warner Bros.'
```

```python
s_companies = data['production_companies'].str.split('|',expand=True).stack().unique()
df_result = pd.DataFrame(columns = ['production_companies', 'length'])
df_result['length'] = df_result['length'].astype('int')
for company in s_companies:
    df_company = data[data['production_companies'].str.contains(company,False)]['original_title']
    for original_title in df_company:
        index = len(df_result)
        df_result.loc[index,'production_companies'] = company
        df_result.loc[index,'length']  = len(original_title.split(' '))
df_result.groupby('production_companies')['length'].max().sort_values(ascending=False).head(10)
```
